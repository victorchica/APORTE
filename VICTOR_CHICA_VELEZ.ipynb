Python transforma TRUE a 1 y FALSE a 0. Como hay muchos registros sumamos el resultado de la función is.na(), si el resultado es 1 o mayor es porque hay valores faltantes si el resultado es 0 significa que no hay valores faltantes

GRAFICANDO UNA SERIE DE TIEMPO

```python
#Libreria para gráficar
import matplotlib.pyplot as plt
```

```python
st_datos.Brasil.plot(figsize=(20,5), title = "S&P500 Prices")#figsize(20,5) sirve para darle tamaño mas grande al gráfico
plt.show()#no aparece el mensaje de la parte superior del gráfico
```

```python
st_datos.Italia.plot(figsize=(20,5), title = "DAX30 Prices")
plt.show()
```

```python
st_datos.Colombia.plot(figsize=(20,5), title = "FTSE100 Prices")
plt.show()
```

```python
st_datos.España.plot(figsize=(20,5), title = "NIKKEI225 Prices")
plt.show()
```

```python
st_datos.Brasil.plot(figsize=(20,5), title = "S&P500 Prices")
st_datos.Italia.plot(figsize=(20,5), title = "DAX30 Prices")
st_datos.Colombia.plot(figsize=(20,5), title = "FTSE100 Prices")
st_datos.España.plot(figsize=(20,5), title = "NIKKEI225 Prices")
plt.title("S&P  DAX  FTSE  NIKKEI")
plt.show()
```
